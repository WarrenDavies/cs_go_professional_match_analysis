```python
# Winningest teams by total wins
```

```python
(df_matches_exploded
    .groupBy('team_name')
    .agg(
        F.count('team_name').alias('matches'),
        F.sum(F.col('won')).alias('wins')
    )
    .withColumn('win_perc', F.round(F.col('wins') / F.col('matches'), 2))
    .orderBy(F.col('wins').desc())
    .show(100)
)
```

```python
# Winningest teams by win percent
```

```python
(df_matches_exploded
    .groupBy('team_name')
    .agg(
        F.count('team_name').alias('matches'),
        F.sum(F.col('won')).alias('wins')
    )
    .withColumn('win_perc', F.round(F.col('wins') / F.col('matches'), 2))
    .orderBy(F.col('win_perc').desc())
    .show(100)
)
    
```
